## Task description

*   Create a program that asks the user for their name, favorite food, and number of
meals they eat per day. The program should then calculate the total meals in a
week
*    and display all details clearly to the user.You must provide:
*   Algorithm
*   Pseudocode
*   Python Code implementing the solution



# Algorithm Steps


1.   Read the user’s name.


1.   Read the user’s favorite food

1.   Read the number of
meals he eats  per day as text , then convert it to an integer

1.   calculate the total meals in a week


2.   Display the name, favorite food, the number of
meals he eats  per day,the total meals in a week

# Pseudocode

*   START
*   read name
*   read favorite_food
*   read number_of_meals_txt
*   number_of_meals_perday= int(read number_of_meals_txt)
*   total_meals_inweek=7*number_of_meals_perday
*   print name,favorite_food,number_of_meals_perday,total_meals_inweek
*   END




# Python Code

```python
name=input("Enter your name: ") #get user name
favorite_food=input("Enter your favorite food: ") #get favorite food
number_of_meals_txt=input("Enter the number of meals you eat per day: ")# Get number of meals he eats per day as str
number_of_meals_perday=int(number_of_meals_txt) # convert number_of_meals to integer
total_meals_inweek=7*number_of_meals_perday #Calculate Total meals he eats in week
# Display types of:
print("Type of Name:",type(name))
print("Type of Favorite_Food:",type(favorite_food))
print("Type of Number of Meals:",type(number_of_meals_perday))
# Print results
print("*"*40)
print("*"*10+"Information about user"+"*"*10)
print("User Name:", name)
print("Favorite Food:", favorite_food)
print("Number of meals per day:", number_of_meals_perday)
print("Total meals in a week: ",total_meals_inweek)
print("*"*40)
print("Thanks For You")
```
