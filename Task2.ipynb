# **Task Description**
#  Temperature Checker
## You are given a list of temperature readings.
## You need to:
###1.  Loop through each temperature
###2.  Skip any reading below 0 (invalid) → continue
###3.  Stop completely if the temperature is above 50 (danger) → break
###4.  For every valid reading:
   *  If it’s below 15 → “Cold”
   *  If it’s between 15 and 30 → “Normal”
   *  Otherwise → “Hot”
###5.  condition: check if the number is even or odd
###6.  Inside each iteration, print a small nested-loop pattern
###7.  Use loop else if the loop finishes normally
###8.  After finishing, perform a simple while loop countdown




```python
temperatures = [12, 18, -5, 30, 7, 55, 20]
for i in temperatures:
  print("temperature: ",i)
  #Skip any reading below 0 (invalid)
  if i<0:
    print("The State:  Invalid")
    continue
  #Stop completely if the temperature is above 50 (danger)
  if i>=50:
    print("The State: danger")
    break
  #The state For every valid reading
  if i<15:
    print("The State: Cold")
  elif i>=15 and i<=30:
    print("The State: Normal")
  else:
    print("The State: Hot")
  # check if the number is even or odd
  if i%2==0:
    print("Type: Even")
  else:
    print("Type: Odd")
  # nested-loop pattern
  print("Pattern")
  for x in range(4):
    for j in range(3):
      print("*",end="")
    print()
else:
  print("The loop finished")
#while loop countdown
print("Countdown:")
count=4
while count>0:
  print(count)
  count-=1
else:
  print("Done")
```
