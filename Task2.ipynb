```python
temperatures = [12, 18, -5, 30, 7, 55, 20]
for i in temperatures:
  print(i)
  if i<0:
    print("temperature: ",i,"Invalid")
    continue
  if i>=50:
    print("temperature: ",i,"danger")
    break
  if i<15:
    print("Cold")
  elif i>=15 and i<=30:
    print("Normal")
  else:
    print("Hot")
  if i%2==0:
    print("Even")
  else:
    print("Odd")
  print("Pattern")
  for x in range(4):
    for j in range(3):
      print("*",end="")
    print()
else:
  print("The loop finished")
#condition
count=4
while count>0:
  print(count)
  count-=1
else:
  print("Done")
```
